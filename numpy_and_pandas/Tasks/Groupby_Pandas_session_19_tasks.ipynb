### `Q-3:` Make groups based on `"Embarked"` column. And for each of this embarked group, make another group based on `"Pclass"` and find out the average fare (round off up to 2 decimal places) for each "Pclass" for each group of "Embarked".

### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
def z_normalize(group, L):
  
  for col in L:
    group['zscore_'+col] = (group[col] - group[col].mean())/group[col].std()
  return group
````

> Chasing mean team batting in second inning

````python
ipl[(ipl.innings == 2)].groupby(['ID','batter'])['batsman_run'].sum().sort_values(ascending=False).head()
#(1).values[0]
#.groupby('batter')['batsman_run'].sum().sort_values(ascending=False)
#.head(1).values[0]
````

````python
balls_100 = ipl[ipl.innings == 2].groupby('batter')['ballnumber'].count()
batter_100_l = balls_100[balls_100.gt(100)].index.tolist()
SR = ipl[ipl.batter.isin(batter_100_l) & (ipl.innings == 2)].groupby('batter').agg({'ballnumber':'count',
                                                                              'batsman_run':'sum'}).reset_index()
SR['Strike_Rate'] = SR['batsman_run']/SR['ballnumber'] * 100
SR.sort_values('Strike_Rate', ascending = False).head()
````

###`Q-6` Most Successful bowler against any batsman. Find that pair of bowler and batsman.
> Most Successful in terms of dissmissal. A bowler who have dissmissed any batsman most no of times. If any two pairs have same no of dissmisal, consider runs conceded by bowler to that batsman. Those who have concede lesser runs is more successful.

````python
# code here
#grp = ipl[ipl.player_out.notnull()].groupby(['player_out','bowler']).agg('')['ID'].count().reset_index().set_axis(['Batter','bowler','No_of_Dismissal'], axis=1)
ipl["IsBatterOut"] = ipl.batter == ipl.player_out
grp = ipl.groupby(['batter','bowler']).agg({'IsBatterOut':'sum',
                                            'batsman_run':'sum'}).reset_index().set_axis(['Batter','bowler','No_of_Dismissal','Total_Runs'], axis=1)
df = grp[grp['No_of_Dismissal'] == grp['No_of_Dismissal'].max()]
df.sort_values('Total_Runs', ascending=True)
````

###`Q-7`: Most successful batting pair in IPL. Batting pair who have scored most runs playing together.


````python
# code here
import numpy as np
def func(x):
  #print('-'.join(np.sort(x.values)))
  return '-'.join(np.sort(x.values))
ipl["batter-pair"] = ipl[["batter", "non-striker"]].apply(func,axis=1)
ipl.groupby(['batter-pair'])['total_run'].sum().sort_values(ascending = False).reset_index().head(1)
````

###`Q-8:` Make a dataframe for all batting pairs played together.
```
Batsman1 Batsman2 Runs Avg StrikeRate
```

> Just to ease this question you can count wide-balls for strike rate.

````python
# code here
ipl
````

````python


df = ipl.groupby(['batter-pair']).agg({'total_run':['sum'],
                                                "isWicketDelivery": "sum",
                                                                       'ballnumber': 'count'}).reset_index()

df['Batsman1'] = df["batter-pair"].apply(lambda x:x.split('-')[0])
df['Batsman2'] = df["batter-pair"].apply(lambda x:x.split('-')[1])
df.columns=['batter-pair','Runs', 'isWicketDelivery','BallCount','Batsman1','Batsman2']


df['Avg'] = df['Runs'] / df['isWicketDelivery']
df
````

````python
df['SR'] = df['Runs']/df['BallCount'] * 100
df.drop(columns=['BallCount','isWicketDelivery','batter-pair'],inplace=True)
df[['Batsman1','Batsman2','Runs',	'Avg','SR']].sort_values('Runs', ascending=False)
````
